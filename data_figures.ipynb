```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataout = np.loadtxt('data.txt')
```

```python
np.shape(dataout)
```

```python
N = int(dataout[0])
dt = dataout[1]
dataout = dataout[2:]
df = 0.5
N, dt
```

```python
data = np.ndarray((int(df/dt) + 1, N))
for i in range((int(df/dt) + 1)):
    data[i] = dataout[i*N:(i+1)*N]
```

```python
x = np.linspace(0, df, N)
```

```python
df/dt
```

```python
for i in range(0, int(df/dt/2), int(df/dt/10)):
    plt.plot(x, data[i])
plt.legend(['t1', 't2', 't3', 't4', 't5'])
plt.show()
```

```python
for i in range(0, int(df/dt)):
    plt.plot(x, data[i])
plt.legend(['t1', 't2', 't3', 't4', 't5'])
plt.show()
```
